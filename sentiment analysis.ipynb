# NLTK Sentiment Analysis (VADER)

This notebook performs sentiment analysis using NLTK's VADER model.

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# Download VADER lexicon if not already available
nltk.download('vader_lexicon')

sia = SentimentIntensityAnalyzer()
```

```python
def analyze_sentiment(text):
    scores = sia.polarity_scores(text)

    compound = scores['compound']
    if compound >= 0.05:
        label = 'Positive'
    elif compound <= -0.05:
        label = 'Negative'
    else:
        label = 'Neutral'

    return scores, label

# Try your own text here
text = 'The product quality is amazing and delivery was super fast!'
scores, label = analyze_sentiment(text)

print('Text:', text)
print('Sentiment:', label)
print('Scores:')
for k, v in scores.items():
    print(f'  {k}: {v}')
```
