# E-commerce Open-Source + AI Augmentation
Load a real open-source e-commerce dataset, show its original shape, add synthetic rows, and save the combined dataset.

```python
from pathlib import Path

import pandas as pd

from augment_ecommerce_with_ai import (
    OPEN_SOURCE_URL,
    generate_augmented_rows,
    load_open_source_ecommerce,
)
```

```python
output_path = Path('ecommerce_open_source_augmented.csv')

base_df = load_open_source_ecommerce()
print('Open-source dataset URL:', OPEN_SOURCE_URL)
print('Original shape before augmentation:', base_df.shape)

synthetic_rows = generate_augmented_rows(base_df=base_df, extra_rows=2500)
print('Synthetic rows shape:', synthetic_rows.shape)

base_tagged = base_df.copy()
base_tagged['dataset_source'] = 'open_source_original'
synthetic_tagged = synthetic_rows.copy()
synthetic_tagged['dataset_source'] = 'ai_augmented_rows'

combined = pd.concat([base_tagged, synthetic_tagged], ignore_index=True)
combined.to_csv(output_path, index=False)

print(f'Saved {output_path} with {len(combined)} rows')
print('Combined shape:', combined.shape)
combined.head()
```
